```python
import numpy as np
import matplotlib.pyplot as plt
from visual.generateData import generate_all_data, sort_by_start

data = generate_all_data(iterations=2, cube_factor=1, cube_size=1)
```

```python
data110 = sort_by_start(data)[(1, 1, 0)]
```

```python
# Convert lists to NumPy arrays
theta_vals = np.array(data110['theta'])  # theta from 0 to 2*pi
phi_vals = np.array(data110['phi'])      # phi from 0 to pi
depth = 2
values = np.array([[float(delta) for delta in l][depth - 1] for l in data110['delta']])

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Scatter plot instead of surface plot
sc = ax.scatter(theta_vals, phi_vals, values, c=values, cmap='viridis')

ax.set_xlabel('Theta')
ax.set_ylabel('Phi')
ax.set_zlabel('Value')

fig.colorbar(sc, shrink=0.5, aspect=5)
plt.show()
```
